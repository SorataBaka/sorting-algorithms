```python
import matplotlib.pyplot as plt
import numpy as np

# Font for Japanese display (macOS)
plt.rcParams['font.family'] = 'Hiragino Sans'

# Data
algorithms = ['Counting Sort', 'Quick Sort', 'Bucket Sort', 'Insertion Sort', 'Shaker Sort', 'Bubble Sort']
datasets = ['data1.dat', 'data2.dat', 'data3.dat']

# Execution times for each algorithm across datasets
times = {
    'Counting Sort': [0.000764, 0.000775, 0.000791],
    'Quick Sort': [0.008333, 0.008439, 0.008298],
    'Bucket Sort': [0.031620, 0.031297, 0.032391],
    'Insertion Sort': [2.984688, 2.990967, 2.970459],
    'Shaker Sort': [12.459189, 12.589956, 12.603889],
    'Bubble Sort': [14.802736, 14.921321, 14.806170]
}

# Set position of bar groups
x = np.arange(len(datasets))
bar_width = 0.12
offsets = np.linspace(-bar_width*2.5, bar_width*2.5, len(algorithms))

# Create the plot
plt.figure(figsize=(12, 7))
for i, algo in enumerate(algorithms):
    plt.bar(x + offsets[i], times[algo], width=bar_width, label=algo)

# Labels and formatting
plt.xticks(x, datasets)
plt.yscale('log')
plt.xlabel('データセット', fontsize=14)
plt.ylabel('実行時間（秒）', fontsize=14)
plt.title('各ソート手法のデータセット別実行時間（対数スケール）', fontsize=16)
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```
